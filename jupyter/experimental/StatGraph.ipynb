```python
import pandas as pd
import os
list_ships = {}
datadir = "../data/azdata/Ships"
for fname in os.listdir(datadir):
    fname = os.path.join(datadir, fname)
    if((not os.path.isfile(fname)) or "template" in fname):
        print("File {:s} ignored.".format(fname))
        continue
    with io.open(fname, "r", encoding="utf-8-sig") as jf:
        sname = os.path.basename(fname)[:-5]
        print("Ship {:s}'s data added.".format(sname))
        list_ships[sname] = json.load(jf)
```

```python
base, lvl100, lvl100ret, lvl120, lvl120ret = [pd.DataFrame() for _ in range(5)]

def countvalid(vals, numtype, ignoreifhigherthan=5):
    invalid = 0
    for v in vals:
        if(v == ""):
            invalid += 1
        else:
            try:
                v = numtype(v)      
            except ValueError:
                invalid += 1
    return ignoreifhigherthan > invalid

def safeconverttype(val, numtype):
    if(val == ''): return None
    return numtype(val)

def statconvert(statdict):
    #print(statdict)
    return {k:(safeconverttype(v, float) if k not in ["armor"] else v) for k, v in statdict.items()}

for ship in list_ships.values():
    common_datadict = {"name": ship["name"], "navy": ship["navy"], "hull": ship["hull"]}
    base = base.append({**common_datadict, **statconvert(ship["stats"]["base"])}, ignore_index=True)
    lvl100 = lvl100.append({**common_datadict, **statconvert(ship["stats"]["100"])}, ignore_index=True)
    lvl120 = lvl120.append({**common_datadict, **statconvert(ship["stats"]["120"])}, ignore_index=True)
    if(countvalid(ship["stats"]["100retrofit"].values(), float, ignoreifhigherthan=5)):
        lvl100ret = lvl100ret.append({**common_datadict, **statconvert(ship["stats"]["100retrofit"])}, ignore_index=True)
        lvl120ret = lvl120ret.append({**common_datadict, **statconvert(ship["stats"]["100retrofit"])}, ignore_index=True)
```

```python
destroyers = lvl100ret[lvl100ret["hull"] == "Destroyer"]

enemydpm = 1000 # currently accounting for crit, weapon base, rng all at once
enemyshellmod = {"Light": 1.0, "Medium": 0.9, "Heavy": 0.5}
enemyacc = 100
enemyluck = 50
leveldiff = 0

def evade_equation(attacker_hit, defender_evade, attacker_luck, defender_luck, level_diff, skill_mod=0.0):
    return max(0.1, 0.1 + attacker_hit / (attacker_hit + defender_evade + 2) 
                        + (attacker_luck - defender_luck + level_diff) / 1000 + skill_mod)
def theoretical_dpm(enemy_dpm, armor_mods, level_diff):
    level_advantage = 1 + level_diff * 0.2
    return enemy_dpm * armor_mods * level_advantage

#destroyers = destroyers.set_index("name")
destroyers["survival"] = destroyers.apply(lambda row: theoretical_dpm(enemydpm, enemyshellmod[row["armor"]], leveldiff) 
                                                  * (1 - evade_equation(enemyacc, row["evasion"], enemyluck, row["luck"], leveldiff)), axis=1)
destroyers = destroyers.sort_values("survival", ascending=False)
destroyers.head()
```
